```python
#analysing data
#merged order report of ABC and sku master of ABC
ABC_data = pd.merge(ABC_order_report_data, ABC_sku_master_data, how='left', on='SKU')
ABC_data.head()
```

```python
#merged ABC_pincodes and courier company invoice
ABC_pincode = ABC_pincode_data .drop_duplicates(subset=['Customer Pincode'])
courier_invoice = courier_invoice_data[['Order ID','Customer Pincode','Type of Shipment']]
pincodes = courier_invoice.merge(ABC_pincode, on = 'Customer Pincode')
pincodes.head()                                 
```

```python
#merged ABC data and pincodes
sku_master = ABC_data.merge(pincodes, on='Order ID')
sku_master.head()
```

```python
sku_master = sku_master.rename(columns={'Order Qty':'Order_Qty','Weight (g)':'Weight_g'})
```

```python
#g_to_kg = '''Total_wt_per_shipment_kg=(Order_Qty*Weight_g)/kg'''
g_to_kg = '''Total_wt_per_shipment_kg=(Order_Qty*Weight_g)/1000'''
sku_master.eval(g_to_kg, inplace=True)
```

```python
sku_master['Type of Shipment'].value_counts()
```

```python
sku_master.head()
```

```python
#assigning weight slab
def weight_slab(weight):
    a = round(weight%1, 1)
    if a == 0.0:
        return weight
    elif a>0.5:
        return int(weight)+1.0
    else:
        return int(weight)+0.5
```

```python
#adding weight slab column
sku_master['Weight_slab_ABC_kg'] = sku_master['Total_wt_per_shipment_kg'].apply(weight_slab)
courier_invoice_data['Weight_slab_courier_kg'] = courier_invoice_data['Total weight by courier'].apply(weight_slab)
```

```python
sku_master.head()
```

```python
courier_invoice_data.head()
```

```python
#expected charges as per ABC
expected_charges = []
for i in range(len(sku_master)):
    fwd = 'fwd_' + sku_master['Zone_ABC'][i]
    fwd_fixed = courier_rates_data[fwd + '_fixed'][0]
    fwd_additional = courier_rates_data[fwd + '_additional'][0]
    rto = 'rto_' + sku_master['Zone_ABC'][i]
    rto_fixed = courier_rates_data[rto + '_fixed'][0]
    rto_additional = courier_rates_data[rto + '_additional'][0]
    if sku_master['Type of Shipment'][i]=='Forward charges':
        if sku_master['Weight_slab_ABC_kg'][i]<0.5:
            expected_charges.append(fwd_fixed)
        else:
            additional_weight = (sku_master['Weight_slab_ABC_kg'][i]-0.5)/0.5
            expected_charges.append(fwd_fixed+additional_weight*fwd_additional)
    if sku_master['Type of Shipment'][i]=='Forward and RTO charges':
        if sku_master['Weight_slab_ABC_kg'][i]<=0.5:
            expected_charges.append(fwd_fixed+rto_fixed)
        else:
            additional_weight = (sku_master['Weight_slab_ABC_kg'][i]-0.5)/0.5
            expected_charges.append(fwd_fixed+additional_weight*(fwd_additional+rto_additional))
sku_master['Expected charge as per ABC(Rs.)'] = expected_charges
```

```python
sku_master.columns
```

```python
ResultABC = sku_master[['Order ID', 'SKU', 'Order_Qty', 'Weight_g', 'Customer Pincode', 'Type of Shipment', 'Warehouse Pincode', 'Zone_ABC',
                        'Total_wt_per_shipment_kg', 'Weight_slab_ABC_kg', 'Expected charge as per ABC(Rs.)']]
ResultABC.head()
```

```python
courier_invoice_data.columns
```

```python
Resultcourier = courier_invoice_data[['AWB Code', 'Order ID', 'Total weight by courier', 'Warehouse Pincode', 'Customer Pincode',
                                      'Zone_courier', 'Type of Shipment', 'Billing Amount (Rs.)', 'Weight_slab_courier_kg']]
Resultcourier.head()
```

```python
finalResult = ResultABC.merge(Resultcourier, on='Order ID', how='left')
finalResult.head()
```

```python
finalResult = finalResult.rename(columns={'Expected charge as per ABC(Rs.)':'Expected_charges_ABC', 'Billing Amount (Rs.)':'Billing_Amount'})                                    
```

```python
#difference between expected charges and billed charges
diff_ec_bc = '''Diff_expected_billed = (Expected_charges_ABC - Billing_Amount)'''
finalResult.eval(diff_ec_bc, inplace=True)
```

```python
finalResult.head()
```

```python
finalResult = finalResult.rename(columns={'AWB Code':'AWB Number', 'Total_wt_per_shipment_kg':'Total weight as per ABC (KG)', 'Order_Qty':'Order Qty', 'Zone _ABC':'Delivery Zone as per ABC','Total weight by courier':
                                          'Total weight as per Courier Company (KG)','Weight_g':'Weight (g)', 'Zone_courier':'Delivery Zone charged by Courier Company','Total_wt_per_shipment_kg':
                                          'Total weight as per ABC (KG)','Weight_slab_ABC_kg':'Weight slab as per ABC (KG)', 'Billing_Amount':'Charges Billed by Courier Company (Rs.)', 'Weight_slab_courier_kg':
                                          'Weight slab charged by Courier Company (KG)', 'Expected_charges_ABC':'Expected charge as per ABC(Rs.)', 'Diff_expected_billed':'Difference Between Expected Charges and Billed Charges (Rs.)'})       
```

```python
finalResult = finalResult.rename(columns={'Zone_ABC':'Delivery Zone as per ABC', 'Expected charge as per ABC(Rs.)':'Expected Charge as per ABC (Rs.)'})
```

```python
finalResult.head()
```

```python
FinalResult = finalResult[['Order ID', 'AWB Number', 'Total weight as per ABC (KG)', 'Weight slab as per ABC (KG)', 'Total weight as per Courier Company (KG)',
                   'Weight slab charged by Courier Company (KG)', 'Delivery Zone as per ABC', 'Delivery Zone charged by Courier Company', 'Expected Charge as per ABC (Rs.)',
                   'Charges Billed by Courier Company (Rs.)', 'Difference Between Expected Charges and Billed Charges (Rs.)']].copy()
```

```python
#summary table
correctly_charged = finalResult[finalResult['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0.0]
equal = ['Total orders where ABC has been correctly charged', len(correctly_charged), sum(correctly_charged['Charges Billed by Courier Company (Rs.)'])]
```

```python
overcharged = finalResult[finalResult['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0.0]
more = ['Total Orders where ABC has been overcharged', len(overcharged), np.absolute(sum(overcharged['Charges Billed by Courier Company (Rs.)']))]
```

```python
undercharged = finalResult[finalResult['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0.0]
less = ['Total Orders where ABC has been undercharged', len(undercharged), np.absolute(sum(undercharged['Charges Billed by Courier Company (Rs.)']))]
```

```python
summary = pd.DataFrame([equal,more,less], columns=['', 'Count', 'Amount (Rs.)'])
```

```python
#create a excel using writer object
with pd.ExcelWriter("Result.xlsx") as writer:
    FinalResult.to_excel(writer, sheet_name='Calculations', index=False)
    summary.to_excel(writer, sheet_name='Summary')
```
